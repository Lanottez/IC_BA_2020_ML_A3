```python
import pandas as pd
import numpy as np
import math 
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import shuffle
import time
```

# Question 1

```python
loandata = pd.read_csv('loandata.csv') 
loandata
```

# Question 2-3

```python
loandata_shuffled = shuffle(loandata)

training_data = loandata_shuffled[0:1000]
validation_data = loandata_shuffled[1000:1500]
test_data = loandata_shuffled[1500:2000]

training_data_x = training_data.drop(columns = ['Default'])
training_data_x_dummy = pd.get_dummies(training_data_x)
training_data_y = training_data.Default

validation_data_x = validation_data.drop(columns = ['Default'])
validation_data_x_dummy = pd.get_dummies(validation_data_x)
validation_data_y = validation_data.Default

test_data_x = test_data.drop(columns = ['Default'])
test_data_x_dummy = pd.get_dummies(test_data_x)
test_data_y = test_data.Default
```

# Question 4

```python
# Is the question ask for counting mode of Default column under validation dataset? It is half/half split???
print("Count of Yes in validation dataset: {}.".format(len(validation_data.Default=="Yes")))
print("Count of No in validation dataset: {}.".format(len(validation_data.Default=="No")))
print('Since the count of yes and no in the validation dataset if half/half split, it is not possible to use naive benchmark (majority predictor) to predict the validationset.')
```

# Question 5 

```python
dt_scores_list_training = []
dt_scores_list_validation = []
training_time = []
best_depth = 0
best_score = 0

# Return the maximum possible depth of this decision tree classifier 
clf = DecisionTreeClassifier() 
clf.fit(training_data_x_dummy,training_data_y)
max_depth = clf.get_depth()
depth_list = range(1,max_depth)

# Run through all possible max depth, and return the max depth the provides the best classifer on predicting validation data
for depth in depth_list:
    clf = DecisionTreeClassifier(max_depth=depth) 
    start = time.time()
    clf.fit(training_data_x_dummy,training_data_y)
    end = time.time()
    dt_scores_list_training.append(clf.score(training_data_x_dummy,training_data_y))
    dt_scores_list_validation.append(clf.score(validation_data_x_dummy,validation_data_y))
    training_time.append(end-start)
    if best_score < clf.score(validation_data_x_dummy,validation_data_y):
        best_score = clf.score(validation_data_x_dummy,validation_data_y)
        best_depth = depth

# Create the data frame to store all the data
dt_scores_list_df = pd.DataFrame(data=dt_scores_list_training,columns=['Score of training data'])
dt_scores_list_df['Score of validation data'] = dt_scores_list_validation
dt_scores_list_df['Training time'] = training_time
dt_scores_list_df['Maximum Depth'] = depth_list

# Return the running time while training the best classifer
max_row = dt_scores_list_df['Score of validation data'].idxmax()
Running_time=dt_scores_list_df.iloc[max_row]['Training time']

# Use the best classifer to predict the training data and validation data
clf = DecisionTreeClassifier(max_depth=best_depth)
clf.fit(training_data_x_dummy,training_data_y)
score_training = clf.score(training_data_x_dummy,training_data_y)
score_validation = clf.score(validation_data_x_dummy,validation_data_y)
print("For predicting the validation data, the maximum depth for the classifier is {},where the best score is {}, and the training time is {:.5f}s.".format(best_depth,best_score,Running_time))        
print('The accuracy rate of using this classifier to predict training data is {}.'.format(score_training))
print('The accuracy rate of using this classifier to predict validation data is {}.'.format(score_validation))


# Plot the maximum depth vs the score of predicting 
plt.subplot(2,1,1)
plt.plot(dt_scores_list_df['Maximum Depth'], dt_scores_list_df['Score of training data'], label = "Validation Data")
plt.plot(dt_scores_list_df['Maximum Depth'], dt_scores_list_df['Score of validation data'], label = "Training Data")
plt.title('Accuracy of Prediction(Decision Tree Tree)')
plt.xlabel('Maximum Depth')
plt.ylabel('Accuracy(%)')
plt.legend()

# Plot the max depth versus the training time
plt.subplot(2, 1, 2)
plt.subplots_adjust(hspace =1,top = 1)
plt.plot(dt_scores_list_df['Maximum Depth'], dt_scores_list_df['Training time'], label = "Training time")
plt.title('Training Time vs The Maximum Depth(Decistion Tree)')
plt.xlabel('Maximum Depth')
plt.ylabel('Training Time(s)')
plt.legend()
```

```python
# Train the decision tree classifier so it returns that highest accuracy rate of predicting training data
alf = DecisionTreeClassifier() # 
alf.fit(training_data_x_dummy,training_data_y)
best_score_training = alf.score(training_data_x_dummy,training_data_y)
best_depth_training = alf.get_depth()

# Use this decision tree classifier to predict the validation data
validation_Score=alf.score(validation_data_x_dummy,validation_data_y)


print('The best max_depth that achieves highest accuracy rate for training data is {}, and the corresponding accuracy rate is {}.'.format(best_depth_training,best_score_training))
print('However, with maximum depth equal to {}, the accuracy rate on predicting validation data accuracy is {} , which is lower than what we have previously. This is due to over-fitting problem.'.format(best_depth_training,validation_Score))
print('In general, the deeper you allow your tree to grow, the more complex your model will become because you will have more splits and it captures more information about the data.')
print('However this means that your model will fit perfectly good with training data, but will not generate good prediction for testing.')
print('Hence we should use validation dataset to determine the proper max_depth of the model.')
```

# Question 6

```python
estimator_list = np.arange(1,50).tolist()
rfs_scores_list_training = []
rfs_scores_list_validation = []
training_time = []

# Run through a list of number of estimators, and return the max depth the provides the best classifer on predicting validation data
for k in estimator_list:
    clf = RandomForestClassifier(n_estimators=k)
    start = time.time()
    clf.fit(training_data_x_dummy, training_data_y)
    end = time.time()
    rfs_scores_list_training.append(clf.score(training_data_x_dummy,training_data_y))
    rfs_scores_list_validation.append(clf.score(validation_data_x_dummy,validation_data_y))
    training_time.append(end-start)
    
# Create the data frame to store all the data
rfs_scores_list_df = pd.DataFrame(data=estimator_list,columns=['Number of Estimator'])
rfs_scores_list_df['Score of training data'] = rfs_scores_list_training
rfs_scores_list_df['Score of validation data'] = rfs_scores_list_validation
rfs_scores_list_df['Training time'] = training_time

# Return the running time while training the best classifer
max_row = rfs_scores_list_df['Score of validation data'].idxmax()
best_number_of_estimator=rfs_scores_list_df.iloc[max_row]['Number of Estimator']
training_time=rfs_scores_list_df.iloc[max_row]['Training time']
best_score = rfs_scores_list_df.iloc[max_row]['Score of validation data']

print("For predicting the validation data, the best number of estimator for the classifier is {:.0f},where the best score is {}, and the training time is {:.5f}s.".format(best_number_of_estimator,best_score,training_time))        


# Plot the number of estimator vs the score of predicting 
plt.plot(rfs_scores_list_df['Number of Estimator'], rfs_scores_list_df['Score of training data'], label = "Training Data")
plt.plot(rfs_scores_list_df['Number of Estimator'], rfs_scores_list_df['Score of validation data'], label = "Validation Data")
plt.title('Accuracy of Prediction(Random Forest Tree)')
plt.xlabel('Number of Estimator')
plt.ylabel('Accuracy(%)')
plt.legend()
```

```python
# Train the decision tree with the best maximum depth, and use it to predict the result of test data
clf = DecisionTreeClassifier(max_depth=best_depth)
start = time.time()
clf.fit(training_data_x_dummy,training_data_y)
score_test = clf.score(test_data_x_dummy,test_data_y)
end = time.time()
Running_time = end - start
print('The accuracy rate of using decision tree to predict test data is {}, and the training time is {:.5f}s.'.format(score_test,Running_time))

# Train the random forest tree with the best number of estimators, and use it to predict the result of test data
clf = RandomForestClassifier(n_estimators=int(best_number_of_estimator))
start = time.time()
clf.fit(training_data_x_dummy,training_data_y)
score_test = clf.score(test_data_x_dummy,test_data_y)
end = time.time()
Running_time = end - start
print('The accuracy rate of using random forest classifier to predict test data is {}, and the training time is {:.5f}s.'.format(score_test,Running_time))
print('While we are trying to predict the defaults in student loan applications, it is better to use decision tree instead of random forest tree.')
print('Using decision tree, we could reach an accuracy of 0.858, and the training time is 0.00802s.')
print('However, using random forest classifier, the accuracy is 0.818, smaller than using decision tree, but the training time is even higher.')
print('Moreover, it is easier to find the max depth while training the decision tree, since there is a function to return the maximum depth while training the model using the training data.')
print('But we have to guess the range of number of estimators for the random forest tree.')
print('Besides, as now we are predict the defaults in student loan application, using decision tree, which has a better interpretability than random forest tree, is preferred.')
```
